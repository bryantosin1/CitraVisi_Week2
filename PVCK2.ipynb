Membuat bentuk Geometri 2D dari OpenCV. Diawali dengan pembuatan black image
dengan tipe data int16.

````python
black_img = np.zeros(shape=(512,512,3),dtype=np.int16)

plt.imshow(black_img)
````

Kemudian menambahkan bentuk persegi panjang sesuai koordinat pt1 dan pt2

````python
cv.rectangle(black_img,pt1=(384,0),pt2=(510,150),color=(0,250,0),thickness=10)
plt.imshow(black_img)
````

Selanjutnya ditambah menambahkan bentuk persegi sesuai koordinat pt1 dan pt2
yang tertulis pada kode program.

````python
cv.rectangle(black_img,pt1=(200,200),pt2=(300,300),color=(0,0,250),thickness=15)
plt.imshow(black_img)
````

Tahap selanjutnya ditambah menambahkan bentuk lingkaran sesuai radius yang
tertulis pada kode program.

````python
cv.circle(black_img,center=(100,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(black_img)
````

Kemudian dilakukan penambahan garis sesuai koordinat pt1 dan pt2 sebagai berikut.

````python
cv.circle(black_img,center=(400,400),radius=10,color=(255,0,0),thickness=8)
plt.imshow(black_img)
````

````python
cv.line(black_img,pt1=(0,0),pt2=(512,512),color=(255,255,255),thickness=5)
plt.imshow(black_img)
````

Penambahan text dengan font yang telah tertulis dengan ukuran yang sudah
ditentukan.

````python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img,text='hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,0),thickness=2,lineType=cv.LINE_AA)
plt.imshow(black_img)
````

1. Perbedaan antara gambar yang ditampilkan tanpa dan dengan Matplotlib:

*   Tanpa Matplotlib: Gambar akan ditampilkan di luar lingkungan Matplotlib Ini biasanya tidak memberikan kontrol yang besar atas tampilan gambar.
*   Dengan Matplotlib:  Ketika  menggunakan Matplotlib untuk menampilkan gambar, juga dapat mengendalikan berbagai aspek tampilan, seperti skala, warna latar belakang, dan label, serta menyimpan gambar dalam berbagai format.

2. Apakah perbedaan dan pengaruhnya pembuatan black image antara tipe data int16 dan int32?

*   Tipe data int16 (int16) Dalam konteks pembuatan gambar hitam, ini berarti setiap piksel gambar akan direpresentasikan dengan 16 bit, yang dapat mengambil nilai dari -32768 hingga 32767. Dengan kata lain, tipe data ini memiliki rentang nilai yang lebih kecil daripada int32.
*   Tipe data int32 (int32) Dalam konteks pembuatan gambar hitam, ini berarti setiap piksel gambar akan direpresentasikan dengan 32 bit, yang dapat mengambil nilai dari -2147483648 hingga 2147483647. Ini memiliki rentang nilai yang lebih luas daripada int16.



*   Pengaruh pembuatan gambar hitam antara keduanya adalah bahwa jika  membuat gambar hitam dengan tipe data int16, gambar akan menggunakan lebih sedikit memori daripada jika  menggunakan int32. Namun, jika  perlu mengolah gambar secara intensif atau melakukan operasi yang memerlukan presisi tinggi, tipe data int32 mungkin lebih cocok karena memiliki rentang nilai yang lebih besar dan presisi yang lebih tinggi.
